```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("/Users/asrithareddy/Desktop/social_media_entertainment_data.csv")
df
```

```python
columns_to_remove = [
    'Average Sleep Time (hrs)', 'Physical Activity Time (hrs)', 'Marital Status','Notifications Received Daily','Daily Music Listening Time (hrs)',
    'Parental Status', 'Tech Savviness Level (scale 1-10)', 'Preferred Device for Entertainment','Ad Interaction Count','Work/Study Time (hrs)','Social Isolation Feeling (scale 1-10)',
    'Data Plan Used', 'Digital Well-being Awareness', 'Sleep Quality (scale 1-10)','News Consumption Time (hrs)','Reading Time (hrs)','Social Media Fatigue Level (scale 1-10)',
    'Social Isolation Feeling (scale 1-10), Daily Entertainment Time (hrs)','Daily Messaging Time (hrs)','Daily Video Content Time (hrs)','Daily Gaming Time (hrs)','Device Type','Internet Speed (Mbps)'
]
df = df.drop(columns=columns_to_remove, errors='ignore')
df
```

```python
df.info()
```

```python
# Check if there are duplicate User IDs
duplicate_user_ids = df['User ID'].duplicated().sum()
print(f"Number of duplicate User IDs: {duplicate_user_ids}")
```

```python
# List of categorical columns
cat_cols = ['Gender', 'Country', 'Primary Platform', 'Occupation', 'Preferred Content Type', 
            'Primary Social Media Goal', 'Preferred Entertainment Platform']

# Check unique values for each categorical column
for col in cat_cols:
    print(f"\nUnique values in {col}: {df[col].unique()}")
```

```python
inconsistent_entries = df[df['Daily Entertainment Time (hrs)'] > df['Screen Time (hrs)']]
print(f"Inconsistent rows: {len(inconsistent_entries)}")
```

```python
df.loc[df['Daily Entertainment Time (hrs)'] > df['Screen Time (hrs)'], 'Daily Entertainment Time (hrs)'] = df['Screen Time (hrs)']
```

```python
print(df.describe())
```

```python
# Define age bins and labels
age_bins = [13, 18, 30, 45, 65]
age_labels = ['Adolescents', 'Young Adults', 'Adults', 'Older Adults']

# Create a new column 'Age Group' based on the bins
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=True)
df
```

```python
#Distribution of Numerical Variables (Histograms)
#Visualize the distribution of numerical features (e.g., social media usage, entertainment spending) to check for skewness, outliers, and central tendencies (mean, median).

num_cols = ['Age', 'Daily Social Media Time (hrs)', 'Daily Entertainment Time (hrs)',
            'Screen Time (hrs)', 'Monthly Income (USD)', 'Monthly Expenditure on Entertainment (USD)']

plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols):
    plt.subplot(2, 3, i+1)
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
for col in cat_cols:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())
```

```python
#Distribution of Categorical Variables (Count Plots)
#Visualize the count of each category in categorical columns to understand the frequency of values (e.g., most popular social media platforms, occupations).

plt.figure(figsize=(15, 10))
for i, col in enumerate(cat_cols):
    plt.subplot(3, 3, i+1)
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, hue=col, palette='pastel', legend=False)
    plt.xticks(rotation=45)
    plt.title(f'Count of {col}')
plt.tight_layout()
plt.show()
```
